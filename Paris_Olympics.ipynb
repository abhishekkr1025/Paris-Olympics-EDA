```python
df['birth_date'].head()
```

```python
df['birth_date'] = pd.to_datetime(df['birth_date'])
```

```python
def calculate_age(birth_date):
    today = datetime.today()
    age = today.year - birth_date.year - ((today.month, today.day) < (birth_date.month, birth_date.day))
    return age
```

```python
df['age'] = df['birth_date'].apply(calculate_age)
```

```python
sns.histplot(df['age'], bins=10)
```

```python
sns.kdeplot(df['age'])
```

```python
schedule['start_date'] = pd.to_datetime(schedule['start_date'])
schedule['end_date'] = pd.to_datetime(schedule['end_date'])
schedule['date'] = schedule['start_date'].dt.date
schedule['date']
```

```python
event_counts = schedule.groupby('date').size().reset_index(name = 'event_count')
event_counts
```

```python
sns.lineplot(data=event_counts, x='date', y='event_count')
plt.xticks(rotation=90)
```

```python
prime_time_start = pd.to_datetime('19:00:00').time()
prime_time_end = pd.to_datetime('21:00:00').time()
schedule['start_time'] = pd.to_datetime(schedule['start_date']).dt.time
prime_time_events = schedule[(schedule['start_time'] >= prime_time_start) & (schedule['start_time'] <= prime_time_end)]
```

```python
prime_time_events.head()
```

```python
prime_time_by_discipline = prime_time_events['discipline'].value_counts()
prime_time_by_discipline
```

```python
prime_time_by_day = prime_time_events['date'].value_counts()
prime_time_by_day
```
